> After investgated the dataset we found the highest reason that seems to let employee to take attrition decision is Job role so we decided to apply A/B Testing to just make sure from the reason's strenth
>
> we assume that:
>
> H0(Null Hypothesis) : There's no a significant relation between job role and attrition 
>
> H1(Alternate Hypothesis) : There's a significant relation between job role and attrition
>
> p-value <= alpha , we will accept the null hypothesis and reject the alternate hypothesis
>
> p-value > alpha , we will reject the null hypothesis and accept the alternate hypothesis
>
> Since the error rate is 5% the alpha will be 0.05
